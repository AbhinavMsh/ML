```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Xblue = [3,3.5,3.7,3.4,3.5]
Yblue = [4.9,4.4,4.9,4.7,4.8]

Xred = [4,4.2,4.5,3.8,3.5]
Yred = [3,2.8,2.6,2.4,2]

pi_x= np.linspace(0,8,10)
pi_y= np.full(10, 3.8)

pi_x1= np.linspace(0,8,10)
pi_y1= np.full(10, 4.4)

pi_x2= np.linspace(0,8,10)
pi_y2= np.full(10, 3)
```

```python
plt.plot(Xblue,Yblue,'bx')
plt.plot(Xred,Yred,'rx')
plt.plot(pi_x, pi_y, color='black', alpha=0.2)
plt.plot(pi_x1,pi_y1,'b--',alpha=0.2)
plt.plot(pi_x2,pi_y2,'r--',alpha=0.2)
plt.plot(np.linspace(0, 4,10), np.linspace(0, 3,10), color='black', linestyle=':',)
plt.plot(np.linspace(0, 3.5,10), np.linspace(0, 4.4,10), color='black', linestyle=':')
plt.plot(np.linspace(3.5,4,10), np.linspace(4.4,3,10), color='green', linestyle='-',label='x2-x1')
plt.plot(np.full(10,4), np.linspace(3,4.4,10), color='black', linestyle='-',label='w vector')


plt.legend()
plt.xlim(0,6)
plt.ylim(0,6)


```
