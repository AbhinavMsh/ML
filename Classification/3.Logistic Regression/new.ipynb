````python
import numpy as np
x = np.array([1, 2, 3])
y = np.array([0, 0, 1])
````

````python
# Initial values
m=0
c=0
alp = 0.1
itr = 100000 # iterations
m_lst=[]
c_lst=[]
cost_lst=[]

sigmod = lambda val:1/(1+ np.exp(-val))

for i in range(itr):
    z = m * x + c 

    y_hat = np.array([sigmod(val) for val in z])

    error = y_hat - y

    new_m = (1/len(x)) *(np.sum(error*x))
    new_c = (1/len(x)) *(np.sum(error))

    m= m-alp * new_m # updating m
    m_lst.append(m)

    c = c - alp * new_c # updating c
    c_lst.append(c)

    cost = -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat)) # cost fnc
    cost_lst.append(cost)

m_lst = np.array(m_lst)
c_lst = np.array(c_lst)
cost_lst = np.array(cost_lst)
````
